```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
x = np.linspace(-10, 10, 100)

plt.plot(x, sigmoid(x),'x')
plt.show()
```

```python
## algorithm of gradient descent

from sklearn.linear_model import SGDRegressor

model = SGDRegressor(max_iter=1000000, tol=1e-5, penalty=None, eta0=0.01, random_state=0)
x = np.linspace(-10, 10, 100)
y = sigmoid(x)


model.fit(x[:, np.newaxis], y)


```

```python
model.predict([[0]])
```

```python
plt.plot(x, y)
plt.plot(x, model.predict(x[:, np.newaxis]))
```
